```python
import handin_03
import sklearn.linear_model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#1.	Perform linear regression where x = bmi and y = bp

xs, ys = handin_03.prepare_data()

xs = np.array(xs).reshape(-1, 1)

model = sklearn.linear_model.LinearRegression()
model.fit(xs, ys)

print('Coefficient: ' + str(model.coef_))
print('Intercept: ' + str(model.intercept_))
#2.	Show the regression equation (y=ax+b)
y = float(model.coef_)*1 + float(model.intercept_)
print('Y = ' + str(y) + ' when X=1')

plt.scatter(xs, ys)
plt.plot(xs, float(model.coef_)*xs + float(model.intercept_))
#3.	Show the regression line on a scatterplot with the other datapoints.

#4.	According to your model, what is y when x=2."
```
