```python
import pandas as pd
import matplotlib.pyplot as plt
import ex1

unmarried = ex1.marrigal_status_changes()[0].to_dict()
married = ex1.marrigal_status_changes()[1].to_dict()
widows = ex1.marrigal_status_changes()[2].to_dict()
divorced = ex1.marrigal_status_changes()[3].to_dict()

keys = list(unmarried.keys())

unmarried_values = list(unmarried.values())
married_values = list(married.values())
widows_values = list(widows.values())
divorced_values = list(divorced.values())

plt.plot(keys, unmarried_values, label = "Unmarried")
plt.plot(keys, married_values, label = "Married")
plt.plot(keys, widows_values, label = "Widows")
plt.plot(keys, divorced_values, label = "Divorced")
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
plt.ylabel('Citizens in millions')
plt.xlabel('Years')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ex1

unmarried = ex1.marriage_by_age()[0].to_dict()
married = ex1.marriage_by_age()[1].to_dict()

keys = list(unmarried.keys())

unmarried_values = list(unmarried.values())
married_values = list(married.values())

plt.plot(keys, unmarried_values, label = "Unmarried")
plt.plot(keys, married_values, label = "Married")
plt.xticks(np.arange(min(keys), max(keys)+1, 10.0))
plt.ylabel('Citizens')
plt.xlabel('Age')
plt.legend()
plt.show()
```
