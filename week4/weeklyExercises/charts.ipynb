```python
import ex1
from matplotlib import pyplot as plt

people_area_dict = ex1.people_by_area()

sorted_dict = {k: v for k, v in sorted(people_are_dict.items(), key=lambda item: item[1])}

values = sorted_dict.values()
keys = sorted_dict.keys()
keys_string = [str(int) for int in keys]

plt.bar(keys_string, values, width=0.5, linewidth=0, align="center")
plt.title('Population by Area', fontsize=12)
plt.xlabel('Area Code')
plt.ylabel('Population')
plt.show()
```
